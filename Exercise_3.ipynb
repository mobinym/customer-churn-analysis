````python
import matplotlib.pyplot as plt
import seaborn as sns
import math

def plot_categorical_bars(df,
                          categorical_columns,
                          rot=45,
                          bar_width=0.5,
                          palette='dark',
                          figsize_per_plot=(6, 4)): 

    n_cols = 3
    n_rows = math.ceil(len(categorical_columns) / n_cols)

    total_figsize = (figsize_per_plot[0] * n_cols, figsize_per_plot[1] * n_rows)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=total_figsize)
    axes = axes.flatten()

    for idx, col in enumerate(categorical_columns):
        ax = axes[idx]
        order = df[col].value_counts().index
        sns.countplot(data=df, x=col, order=order, palette=palette, width=bar_width, ax=ax)
        ax.set_title(col, fontsize=12)
        ax.set_xlabel('')
        ax.set_ylabel('Count')
        ax.tick_params(axis='x', rotation=rot)

    # Ù¾Ø§Ú©â€ŒØ³Ø§Ø²ÛŒ subplotÙ‡Ø§ÛŒ Ø§Ø¶Ø§ÙÙ‡
    for j in range(len(categorical_columns), len(axes)):
        fig.delaxes(axes[j])

    fig.tight_layout(pad=3)
    fig.suptitle('Barplot of Categorical Features', fontsize=20, y=1.02)
    plt.subplots_adjust(top=0.95)
    plt.show()
````

````python
baseline_models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
}
````

````python
results = []

for model_name, model in baseline_models.items():
    print(f"Training and evaluating model: {model_name}")
    
    clf = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    
    clf.fit(X_train, y_train)
    
    y_train_pred = clf.predict(X_train)
    y_test_pred = clf.predict(X_test)
    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    results.append({
        'Model': model_name,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy
    })
    
    print(f"Classification Report (Test) for {model_name}:\n")
    print(classification_report(y_test, y_test_pred))
    print("="*50)

results_df = pd.DataFrame(results)
print("\nSummary of Baseline Model Performance:")
print(results_df)
````
